误差分析：

二阶导数的中心差分公式为：$$f''(x)=\frac{f(x+h)-2f(x)+f(x-h)}{h^2}$$
$f(x \pm h)$在$x$处的Taylor展开为：
$$f(x+h)=f(x)+hf'(x)+\frac{h^2 f''(x)}{2}+\frac{h^3 f^{(3)}(x)}{6}+o(h^4) $$
$$f(x-h)=f(x)-hf'(x)+\frac{h^2 f''(x)}{2}-\frac{h^3 f^{(3)}(x)}{6}+o(h^4) $$
组合成上面的中心差分公式的形式可以得到：
$$f''(x)=\frac{f(x+h)-2f(x)+f(x-h)}{h^2}-\frac{h^2}{12}f^{(4)}(x)+o(h^4)$$

可以发现，这个式子中的第一项是中心差分的精确值，第二项是截断误差项$\varepsilon_a$。误差与$h^2$成正比。

截断误差为：

$$\varepsilon_a=\frac{h^2}{12}f^{(4)}(\xi)$$

其中，$\xi$在$x-h$和$x+h$之间。

假设$f(x+h)$ 和 $f(x-h)$ 的绝对误差均为机器精度 $\delta$，则舍入误差为：

$$\varepsilon_r=\frac{\delta}{h^2}$$

总误差

$$\varepsilon=\varepsilon_a+\varepsilon_r=\frac{h^2}{12}f^{(4)}(\xi)+o(h^4)+\frac{\delta}{h^2}$$

当误差取最小值时，截断误差和舍入误差的和的导数应为0，即

$$\frac{d\varepsilon}{dh}=\frac{1}{3}hf^{(4)}(\xi)-\frac{2\delta}{h^3}=0$$

$$h_{opt} = \sqrt[4]{\frac{6\delta}{|f^{(4)}(\xi)|}}$$

此时,$\varepsilon_{opt}$取到最小值。

$$\varepsilon_{opt}=2\sqrt{6|f^{(4)}(\xi)|\delta}$$

对于32位单精度浮点数，使用23位二进制有效数字以及一位符号位和8位指数位，机器精度$\delta=2^{-23} \approx 1.19 \times 10^{-7}$。

如选取的函数为$f(x)=x^5$，在$x=1$点选取的步长应为$h_{opt}=\sqrt[4]{\frac{2^{-23}}{20}} \approx 0.0087$，误差约为$\varepsilon_{opt}=2\sqrt{720 \times 2^{-23}} \approx 0.019$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1 + 0.5*np.tanh(2*x)

def analytic_derivative(x):
    return 1 - np.tanh(2*x)**2

def central_difference(f, x, h):
    return (f(x + h/2) - f(x - h/2))/h


N = 100
x = np.linspace(-2, 2, N)
h = x[1] - x[0]
df_num = central_difference(f, x, h)

df_ana = analytic_derivative(x)

plt.plot(x, df_ana, label='Analytic solution')
plt.plot(x, df_num, 'o', label='Numerical solution')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return np.exp(-x**2)

def romberg(f, a, b, eps, MAX_ITER):
    T = np.zeros((MAX_ITER, MAX_ITER))

    h = b - a
    T[0, 0] = (f(a) + f(b)) * h / 2

    for i in range(1, MAX_ITER):
        h /= 2
        
        sum_f = 0
        for j in range(1, 2**(i-1) + 1):
            sum_f += f(a + (2*j - 1) * h)
        T[i, 0] = T[i-1, 0]/2 + h * sum_f

        for j in range(1, i+1):
            T[i, j] = (4**j * T[i, j-1] - T[i-1, j-1]) / (4**j - 1)

        if abs(T[i, i] - T[i-1, i-1]) < eps:
            return T[i, i]
        


x_list = np.arange(0, 3.1, 0.1)
E_list=[]
a = 0
eps = 1e-6
MAX_ITER = 20
for x in x_list:
    E = romberg(f, a, x, eps, MAX_ITER)
    E_list.append(E)
    
    print(x,E)
```

```python
plt.plot(x_list,E_list)

plt.title("Line Plot")
plt.xlabel("x")
plt.ylabel("E")
plt.show()
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def MC(N):
    s=0
    for i in range(N):
        x_list=[]
        for d in range(10):
         x_list.append(random.uniform(0,1))
        s+=sum(x_list)**2
    I=s/N
    return I


MC(160000)
```

(c)
对于一般数值求积方法，为了保证误差，随着维数的增加，计算量会随着维数的幂次方而增加。
蒙特卡洛法求定积分的误差可以用中心极限定理来估计。根据中心极限定理，当样本数量足够大时，样本均值的分布趋近于正态分布。因此，可以将蒙特卡洛法的误差看作样本均值的标准差，即：
$$error \approx \frac{\sigma}{\sqrt{N}}$$
其中，$\sigma$ 是被积函数在积分区间上的方差，$N$ 是采样点的数量。由此可以看出蒙特卡洛法的误差随着采样点数量的增加而减小，并且与维数无关。
