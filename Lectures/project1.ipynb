## Project 1: Visualizing Electric Fields


````python
import numpy as np 
import matplotlib.pyplot as plt 
# from math import sqrt, sin, cos 

# parameters
charge = {1: (-1, 0), -1: (1, 0)}    # dictionary: positive and negative charges --> positions
N = 51     # No. of spatial grid in x/y axis
L = 2.     # 2D regime [-L, L]
xs = np.linspace(-L, L, N)
ys = np.linspace(-L, L, N)

# Initialize the values of electric fields 
Ex = np.zeros((N, N))
Ey = np.zeros((N, N))

for i, x in enumerate(xs):
    for j, y in enumerate(ys):
        for q, pos in charge.items():
            x0, y0 = pos 
            R = np.sqrt((x - x0)**2 + (y - y0)**0 )
            Ex[i][j] += q * (x - x0) / R**3
            Ey[i][j] += q * (y - y0) / R**3


plt.figure(figsize=(6, 6))
plt.streamplot(xs, ys, Ex, Ey, density=1.5, color='k')
plt.xlabel('$x$')
plt.ylabel('$y$')     
plt.scatter(charge[-1][0], charge[-1][1], s=100, c='b')
plt.scatter(charge[1][0], charge[1][1], s=100, c='r')
plt.title('Electric Field Distribution of Several Charges')
plt.show()
````

There are two key tasks in this project:

- Obtain the Ex/Ey distributions from Coulomb law
- Plot the electric field lines

Each task can be written into one function.

````python
import numpy as np 
import matplotlib.pyplot as plt 
# from math import sqrt, sin, cos 

# FUNCTION 1:
def makefield(charges, xs, ys):
    N = len(xs)
    Ex = np.zeros((N, N))
    Ey = np.zeros((N, N))

    for i, x in enumerate(xs):
        for j, y in enumerate(ys):
            for q, pos in charges.items():
                x0, y0 = pos 
                R = np.sqrt((x - x0)**2 + (y - y0)**0 )
                Ex[i][j] += q * (x - x0) / R**3
                Ey[i][j] += q * (y - y0) / R**3
    
    return Ex, Ey

#---------------- MAIN PROGRAM --------------------#
# parameters
charges = {1: (-1, 0), -1: (1, 0)}    # dictionary: positive and negative charges --> positions
N = 51     # No. of spatial grid in x/y axis
L = 2.     # 2D regime [-L, L]
xs = np.linspace(-L, L, N)
ys = np.linspace(-L, L, N)

Ex, Ey = makefield(charges, xs, ys)

plt.figure(figsize=(6, 6))
plt.streamplot(xs, ys, Ex, Ey, density=1.5, color='k')
plt.xlabel('$x$')
plt.ylabel('$y$')     
plt.scatter(charge[-1][0], charge[-1][1], s=100, c='b')
plt.scatter(charge[1][0], charge[1][1], s=100, c='r')
plt.title('Electric Field Distribution of Several Charges')
plt.show()
````
