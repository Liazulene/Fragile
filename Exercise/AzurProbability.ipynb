```python
import random
import numpy as np
from matplotlib import pyplot as plt 
import time


# a=环，b=Luna，c=玛丽萝丝，d=穗香，e=霞，f=凪咲，g=莫妮卡，h=海咲
prizes = ["a", "b", "c", "d", "e", "f", "g","h","others"]

# 概率列表
probabilitiesA = [0.020, 0.020, 0.005, 0.005, 0.005, 0.025, 0.025, 0.005, 0.890]
probabilitiesB = [0.000, 0.000, 0.020, 0.020, 0.020, 0.025, 0.025, 0.005, 0.885]

i_list = []
cir = 0

while cir <= 100000:

    i=0
    count = np.zeros(9)
    results = []

    while count[0] == 0 or count[1] == 0 :
        result = random.choices(prizes, probabilitiesA)[0]
        count[prizes.index(result)] += 1
        results.append(result)
        i += 1 
        


    while min(count[2],1) + min(count[3],1) + min(count[4],1) + min(count[5],1) + min(count[6],1) <= 3:
        result = random.choices(prizes, probabilitiesB)[0]
        count[prizes.index(result)] += 1
        results.append(result)
        i += 1 

    counts = {}
    for prize in prizes:
        counts[prize] = results.count(prize)

    i_list.append(i)
    cir += 1

'''
for prize, count in counts.items():
    print(f"{prize}: {count} ")

print(f"总抽数：{i}")
'''
counts = {}
for num in i_list:
    if num in counts:
        counts[num] += 1
    else:
        counts[num] = 1

sorted_counts = sorted(counts.items())

total = len(i_list)
cumulative_counts = 0
cumulative_probabilities = {}
for num, count in sorted_counts:
    cumulative_counts += count
    cumulative_probabilities[num] = cumulative_counts / total

max_num = 1000  
choukacishu=[]
gailv=[]
for num, prob in cumulative_probabilities.items():
    choukacishu.append(num)
    gailv.append(prob)
    if num > max_num:
        break
    print(f"{num}: {prob:.4f}")


plt.plot(choukacishu,gailv)
plt.rcParams['font.sans-serif']=['SimHei']
plt.xlabel(f"抽卡次数")
plt.ylabel(f"毕业概率")
plt.xlim(0, choukacishu[-1])
plt.ylim(0, 1)
plt.grid(True)
plt.show
```

```python
import random
import numpy as np
import time

random.seed(time.time())
# 奖品列表
# a=环，b=Luna，c=玛丽萝丝，d=穗香，e=霞，f=凪咲，g=莫妮卡，h=海咲
prizes = ["a", "b", "c", "d", "e", "f", "g","h","others"]

# 奖品概率列表
probabilitiesA = [0.02, 0.02, 0.005, 0.005, 0.005, 0.025, 0.025, 0.005, 0.890]
probabilitiesB = [0, 0, 0.02, 0.02, 0.02, 0.025, 0.025, 0.005, 0.885]

i=0
count = np.zeros(9)
results = []

while count[0] == 0 or count[1] == 0:
        result = random.choices(prizes, probabilitiesA)[0]
        count[prizes.index(result)] += 1
        results.append(result)
        i += 1 


while min(count[2],1) + min(count[3],1) + min(count[4],1) + min(count[5],1) + min(count[6],1) <= 3:
        result = random.choices(prizes, probabilitiesB)[0]
        count[prizes.index(result)] += 1
        results.append(result)
        i += 1 

counts = {}
for prize in prizes:
        counts[prize] = results.count(prize)
for prize, count in counts.items():
    print(f"{prize}: {count} ")

print(f"总抽数：{i}")
```
